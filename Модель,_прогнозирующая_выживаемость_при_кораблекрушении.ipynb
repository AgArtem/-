```
import torch
import torch.nn as nn
import torch.optim as optim
import torch.optim.lr_scheduler as lr_scheduler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train_tensor = torch.tensor(X_train_copy.to_numpy(), dtype=torch.float32)
y_train_tensor = torch.tensor(y2_train.to_numpy(), dtype=torch.long)
X_test_tensor = torch.tensor(X_test_copy.to_numpy(), dtype=torch.float32)
y_test_tensor = torch.tensor(y2_test.to_numpy(), dtype=torch.long)

class Net(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_classes):
        super().__init__()
        layers = [nn.Linear(input_dim, hidden_dim), nn.ReLU(), nn.BatchNorm1d(hidden_dim)]
        for _ in range(num_layers - 1):
            layers.extend([nn.Linear(hidden_dim, hidden_dim), nn.ReLU(), nn.BatchNorm1d(hidden_dim)])
        layers.append(nn.Linear(hidden_dim, num_classes))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

def objective(trial):
    hidden_dim = trial.suggest_int('hidden_dim', 16, 128)
    num_layers = trial.suggest_int('num_layers', 1, 7)
    lr = trial.suggest_float('lr', 1e-6, 1e-1, log=True)
    gam = trial.suggest_float('gamma', 1e-6, 1)

    model = Net(input_dim=X_train_tensor.shape[1], hidden_dim=hidden_dim, num_layers=num_layers, num_classes=num_classes)
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss().type(torch.int)
    scheduler = lr_scheduler.ExponentialLR(optimizer, gamma=gam, last_epoch=-1)

    model.train()
    for epoch in range(50):
        optimizer.zero_grad()
        outputs = model(X_train_tensor)
        loss = criterion(outputs, y_train_tensor.long())
        loss.backward()
        optimizer.step()

    scheduler.step()
    model.eval()
    with torch.no_grad():
        preds = model(X_test_tensor)
        predicted_classes = torch.argmax(preds, dim=1).numpy()
    accuracy = accuracy_score(y_test_tensor, predicted_classes)
    return accuracy

num_classes = len(set(y2_train))  # количество классов

import optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1000)

print("Лучшие параметры:", study.best_params)
print("Лучшая точность:", study.best_value)
```

Нейросеть справилась лучше, чем классические модели. Это лишь конкретный случай. Если имеем маленький датасет, то нейросеть маловероятно нормально обучить.

```
import torch.optim.lr_scheduler as lr_scheduler

CUDA_DEVICE_ID = 0  # change if needed
device = (
    torch.device(f"cuda:{CUDA_DEVICE_ID}") if torch.cuda.is_available() else torch.device("cpu")
)

model = Net(input_dim=X_train_tensor.shape[1], hidden_dim=128, num_layers=1, num_classes=num_classes)
model.to(device)

num_epochs = 1000
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01013779340053823)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.6834146667092419, last_epoch=-1)

def get_accuracy(model, data_loader):
    predicted_labels = []
    real_labels = []
    model.eval()
    with torch.no_grad():
        y_predicted = model(data_loader[0].to(device))
        predicted_labels.append(y_predicted.argmax(dim=1).cpu())
        real_labels.append(data_loader[1])

    predicted_labels = torch.cat(predicted_labels)
    real_labels = torch.cat(real_labels)
    accuracy_score = (predicted_labels == real_labels).type(torch.FloatTensor).mean()
    return accuracy_score

def validate(model, data_loader, criterion):
    model.eval()
    total_loss = 0

    with torch.no_grad():
        x, y = data_loader[0].to(device), data_loader[1].to(device)
        prediction = model(x)
        loss = criterion(prediction, y)
        total_loss += loss.item()
    return total_loss / len(data_loader)

def train_model(model, train_data_loader, test_data_loader, criterion, optimizer, num_epochs, scheduler):
    loss_history = []
    train_history = []
    test_history = []

    for epoch in range(num_epochs):
        model.train()
        total_loss = 0

        x, y = train_data_loader[0].to(device), train_data_loader[1].to(device)
        prediction = model(x)
        loss = criterion(prediction, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        ave_loss = total_loss / len(train_data_loader)
        train_accuracy = get_accuracy(model, train_data_loader)
        test_accuracy = get_accuracy(model, test_data_loader)
        val_loss = validate(model, train_data_loader, criterion)
        scheduler.step()

        loss_history.append(ave_loss)
        train_history.append(train_accuracy.item())
        test_history.append(test_accuracy.item())

        print("Epoch: %i, Average loss: %f, Val loss: %f, Train accuracy: %f, Test accuracy: %f" % (epoch+1, ave_loss, val_loss, train_accuracy, test_accuracy))

    return loss_history, train_history, test_history

loss_history, train_history, test_history = train_model(model, (X_train_tensor, y_train_tensor), (X_test_tensor, y_test_tensor), criterion, optimizer, num_epochs, scheduler)
```
